```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#打开文件并读取第一行
origin_data = open("D:/input/data.txt", "r", encoding='utf-8')
line = origin_data.readline()
line
```

```python
# 创建空列表
title_list = []
while line:
    line=line.strip()
    prods=line.split(',')
    for prod in prods:
        title_list.append(prod)
    line=origin_data.readline()
```

```python
title_list=set(title_list)
```

```python
#查看去重后的列表
title_list
```

```python
print("共有", len(title_list) ,"种商品")
```

```python
#关闭文件
origin_data.close()
```

```python
# 创建一个空的dataframe用来存放处理后的数据集
title_list=list(title_list)#coulmns似乎不能是set
changed_data = pd.DataFrame(columns=title_list)
changed_data
```

```python
#再次打开文件
origin_data = open("D:/input/data.txt", "r", encoding='utf-8')
line = origin_data.readline()
while line:
    customer={}
    line=line.strip()
    prods=line.split(',')
    for prod in title_list:
        if prods.__contains__(prod):
            customer[prod]=1
        else:
            customer[prod]=0
    changed_data = changed_data.append(customer, ignore_index=True)
    #changed_data = pd.concat([changed_data, pd.DataFrame(customer, index=[0])], ignore_index=True)
    line=origin_data.readline()
changed_data.head(5)
```

```python
changed_data.to_excel("D:/output/changed_data.xlsx", index=False)
```

```python
frequent_itemsets = apriori(changed_data, min_support=0.05, use_colnames=True)
# 输出频繁项集
frequent_itemsets
```
